```python
Session = sessionmaker(bind=engine)
session = Session()

# Crear un rol si no existe
role = session.scalars(select(Role).where(Role.name == "user")).first()
if not role:
    role = Role(name="user", description="Usuario normal")
    session.add(role)
    session.commit()

# Crear 5 usuarios si no existen
users_data = [
    {"username": "user1", "email": "user1@example.com", "password": "pass1"},
    {"username": "user2", "email": "user2@example.com", "password": "pass2"},
    {"username": "user3", "email": "user3@example.com", "password": "pass3"},
    {"username": "user4", "email": "user4@example.com", "password": "pass4"},
    {"username": "user5", "email": "user5@example.com", "password": "pass5"},
]

for user_data in users_data:
    existing_user = session.scalars(select(User).where(User.email == user_data["email"])).first()
    if not existing_user:
        user = User(
            username=user_data["username"],
            email=user_data["email"],
            password=user_data["password"],
            role_id=role.id
        )
        session.add(user)

session.commit()
```
