```python
from sqlalchemy import create_engine, String, Integer, Date, select, Boolean, ForeignKey,and_,or_, func, select
from sqlalchemy.orm import DeclarativeBase, Mapped, mapped_column, sessionmaker, Session, relationship
from datetime import date

import sys


sys.path.append("../src")


from domain.db import Base,engine
from domain.models import User,Profile,Role,PostTags,Tag,Category,Post,PostFav
Session = sessionmaker(bind=engine)
session = Session()
```

```python
# Query about User-Email and when Created at
```

```python
with Session() as session:
    stmt = (
        select(User)
        .where(User.username ==("user2"))        
        .order_by(User.username.desc())
        )
    
    users_list= session.scalars(stmt).all()
    for user in users_list:
        print(f"Name: {User.username} Email: {User.email} Created at: {User.created_at}")
```

```python
# query about list of user and number of publication they have
```

```python
with Session() as session:
    stmt=(
        select(User.username, func.count(Profile.id).label("Profile_id"))
        .outerjoin(Profile, Profile.user_id == User.id)
        .group_by(User.id, User.username)
    )
    rows = session.execute(stmt)

    for u_name, total in rows:
        print(f"Name: {u_name} Publication: {total}")
```

```python
with Session() as session:
    stmt = (
        select(User)
        .where(and_(
                or_(User.id>2)
                ))
        .order_by(User.id.desc()))
    
    Users= session.scalars(stmt).all()
    for User in Users:
        print(f"Name: {User.username} Email: {User.email} Created at: {User.created_at}")
```
